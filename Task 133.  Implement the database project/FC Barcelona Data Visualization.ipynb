```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example Data: Sponsorship Deals
sponsorship_data = {
    "Sponsor": ["Nike", "Spotify"],
    "AnnualValue (in €M)": [105, 50],
}
```

```python
# Example Data: Financial Summary (2023/2024)
financial_data = {
    "Category": ["Revenues", "Expenses", "Net Profit/Loss"],
    "Amount (in €M)": [325, 350, -25],
}
```

```python
# Convert to DataFrames
sponsorship_df = pd.DataFrame(sponsorship_data)
financial_df = pd.DataFrame(financial_data)
```

```python
# Visualization 1: Sponsorship Contributions
plt.figure(figsize=(8, 5))
plt.bar(sponsorship_df["Sponsor"], sponsorship_df["AnnualValue (in €M)"], color=['blue', 'green'])
plt.title("Sponsorship Contributions (2023/2024)", fontsize=14)
plt.xlabel("Sponsors", fontsize=12)
plt.ylabel("Annual Value (in €M)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("sponsorship_contributions.png")
plt.show()
```

```python
# Visualization 2: Financial Summary (Revenues, Expenses, Net Loss)
plt.figure(figsize=(8, 5))
colors = ['green', 'red', 'gray']
plt.bar(financial_df["Category"], financial_df["Amount (in €M)"], color=colors)
plt.title("Financial Summary (2023/2024)", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Amount (in €M)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("financial_summary.png")
plt.show()
```
